***Busines Objective***

Dataset inii berisi kumpulan data berupa detail transaksi melalui Invoice ID, lokasi penjualan di Branch dan City, informasi pelanggan yang diketahui berdasarkan Customer Type dan Gender, sementara produk dikelompokkan ke dalam kategori Product Line dengan detail Unit Price, Quantity, Tax 5%, dan nilai Total belanja.


**Masalah** yang terdapat pada dataset adalah Perusahaan hanya tahu produk mana yang paling laku, tapi belum tahu produk mana yang paling untung. Produk yang laku keras belum tentu menghasilkan uang banyak kalau selisih harga jual dan modalnya terlalu tipis. Tanpa pemahaman ini, perusahaan berisiko membuang sumber daya untuk memasarkan produk yang keuntungannya sangat tipis atau bahkan rugi.

**Tujuan Analisis** nya yaitu mencari tahu atau identifikasi produk yang menjadi untung besar dan yang hanya menjadi beban stok (untung kecil tapi modal besar) dengan berdasarkan margin laba, serta pola penjualan untuk efisiensi stok dan biaya


---



***Assess Situation***

Untuk kondisi dari Dataset ini adalah, penjualan yang sudah rapi namun belum dimanfaatkan maksimal untuk mencari keuntungan bersih.
**Tantangan** utamanya adalah tidak adanya data waktu (tanggal/jam), sehingga tidak bisa melihat tren penjualan bulanan atau jam sibuk. Maka dari itu, untuk fokus analisis akan dialihkan pada performa kategori produk dan cara pembayaran pelanggan. Dari Dataset ini juga ada risiko angka harga jual dan modal yang tidak sinkron, jadi diperlukan pembersihan data agar hitungan profit tidak salah. Asumsinya efisiensi bisnis akan dinilai berdasarkan perbandingan langsung antara Total_Sales dan Total_Cost yang tersedia di dalam dataset.

***Analytic Goals***

Untuk mengatasi masalah yang sudah dijelaskan, maka diperlukan beberapa langkah yaitu:

1. Menghitung total keuntungan bersih per produk untuk tahu keuntungan tiap produk (Harga - Modal) x Jumlah Terjual.

2. Identifikasi produk yaitu menentukan 10 produk dengan kontribusi profit terbesar dan 10 produk dengan performa terendah untuk di evaluasi.

3. Analisis produk yang tidak efisien, seperti biaya terlalu mendekati harga jual atau produk yang modalnya kemahalan tapi dijual terlalu murah.

4. Visualisasi dengan menampilkan perbandingan dalam bentuk grafik agar mudah dipahami kondisi bisnis nya.
