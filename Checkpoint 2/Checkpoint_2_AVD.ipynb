## **Outliers Values**

Pada tahap ini, melalukan pengecekan nilai ekstrem (outliers) pada kolom yang bertipe data kan numerik. Hal ini dilakukan agar mendapatkan insight yang akurat.

```python
target_cols = ['Unit price', 'Quantity', 'Tax 5%']
results = []

for col in target_cols:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    percent_outliers = (len(outliers) / len(df)) * 100
    results.append({
        'Kolom': col,
        'Jumlah Outliers': len(outliers),
        'Persentase Outliers (%)': round(percent_outliers, 2)
    })
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
display(results_df)
```

* **Unit price**

Dengan presentase 0%

* **Quantity**

Dengan presentase 0%

* **Tax 5%**

Dengan presentase 10.9%

---

Terlihat bahwa data ini memiliki variabilitas yang normal pada harga dan jumlah barang, namun memiliki masalah kualitas yang berat pada pajak. Kehadiran outlier pada kolom Tax 5% wajib ditangani, baik dengan cara menghapus baris anomali tersebut atau menghitung ulang nilainya.

# **EDA (Eksplorasi Data)**

Sebelum masuk ke tahap Empat Pilar, perlu memunculkan variabel keuangan yang belum tersedia secara langsung di dataset, yaitu menghitung Modal (Cost)
Menghitung Sales (Total Penjualan) dan Menghitung Profit (Keuntungan).

```python

df['Modal'] = df['Unit price'] * df['Quantity']
df['Sales'] = df['Modal'] + df['Tax 5%']
df['Profit'] = df['Tax 5%']
display(df[['Product line', 'Modal', 'Sales', 'Profit']].head())
```

## **Comparison/Perbandingan**

Aktivitas: Membandingkan total nominal Product line dan Total profit.

Tujuan: Mengidentifikasi kategori produk mana yang menjadi kontributor keuntungan terbesar dan mana yang memberikan keuntungan paling minim.

Visualisasi: Bar Chart (Grafik Batang).

```python
penjualan_market = df.groupby('Product line')['Profit'].sum().sort_values(ascending=False)

plt.figure(figsize=(10,6))
sns.barplot(x=penjualan_market.index, y=penjualan_market.values, palette='viridis', hue=penjualan_market.index, legend=False)
plt.title('Total Produk Dengan Keuntungan paling besar')
plt.xlabel('Product line')
plt.ylabel('Profit ')
plt.xticks(rotation=45)
plt.show()
```

Kategori inilah yang menjadi kontributor utama pendapatan perusahaan. Produk dalam kategori ini harus dipastikan ketersediaan stoknya agar tidak terjadi kehilangan potensi keuntungan

## **Composition/Komposisi**

Aktivitas: Menganalisis kontribusi profit dari setiap Kategori Produk (Product line) terhadap total keuntungan perusahaan.

Tujuan: Mengetahui kategori produk mana yang memegang porsi paling besar. Hal ini membantu perusahaan menentukan kategori mana yang harus diprioritaskan stoknya karena kontribusinya yang dominan terhadap total profit.

Visualisasi: Pie Chart

```python
sales_by_category = df.groupby('Product line')['Profit'].sum().sort_values(ascending=False)
sales_by_category_top3 = sales_by_category.head(3)

plt.figure(figsize=(10, 8))
sales_by_category_top3.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Proporsi 3 Kategori Produk Teratas Berdasarkan Total Penjualan')
plt.ylabel('')
plt.axis('equal')
plt.show()
```

Grafik Pie Chart menunjukkan kategori produk mana yang paling banyak memberikan keuntungan bagi perusahaan. Visualisasi ini membantu memetakan prioritas bisnis secara instan. Dengan melihat porsi keuntungan setiap kategori, perusahaan dapat lebih bijak dalam mengalokasikan budget pengadaan barang pada kategori yang terbukti memberikan margin laba paling besar.

## **Distribution/Distribusi**

Aktivitas: Memetakan sebaran nilai Profit dan Tax 5% untuk seluruh transaksi.

Tujuan: Mengetahui rentang keuntungan yang paling sering didapat dan secara visual mendeteksi adanya anomali/outlier yang menyimpang dari distribusi data normal.

Visualisasi: Histogram.

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Profit'], bins=20, kde=True)
plt.title('Distribusi (Profit)')
plt.xlabel('Profit')
plt.ylabel('Tax 5%')
plt.show()
```

Visualisasi Histogram membuktikan bahwa meskipun transaksi berada pada rentang profit yang wajar, terdapat outlier yang merusak struktur distribusi data. Ini menjadi dasar kuat untuk melakukan proses penghapusan data anomali.

## **Relationship/Hubungan**

Aktivitas: Menganalisis hubungan antara Modal dengan Profit pada setiap transaksi.

Tujuan: Melihat efisiensi biaya. Mengidentifikasi apakah ada produk yang modalnya terlalu tinggi namun nilai penjualannya rendah, serta mendeteksi ketidaksinkronan harga jual vs modal.

Visualisasi: Heatmap

```python
plt.figure(figsize=(8, 6))
sns.heatmap(data=df[['Modal', 'Profit']].corr(),
            annot=True,
            cmap='viridis',
            fmt='.2f')
plt.title('Korelasi antara Modal dan Profit')
plt.show()
```

Hasil korelasi awal antara Modal dan Profit hanya menunjukkan angka 0,27, yang termasuk dalam kategori korelasi positif yang lemah. Rendahnya angka ini membuktikan adanya anomali data yang sangat kuat (outlier). Oleh karena itu, diperlukan tahap Data Preparation agar hubungan antar variabel keuangan menjadi logis dan akurat.
